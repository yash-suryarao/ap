#### 1. Write a Pandas program to create and display a DataFrame from a specified dictionary data which has the index labels.

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 27, 22, 32, 29],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
}

# Define custom index labels
index_labels = ['A', 'B', 'C', 'D', 'E']

# Create a DataFrame from the dictionary with the specified index labels
df = pd.DataFrame(data, index=index_labels)
print(df)
```

#### 2. Write a Pandas program to get the first 3 rows of a given DataFrame

```python
df = pd.DataFrame(data)

# Get the first 3 rows of the DataFrame
first_3_rows = df.head(3)

print(first_3_rows)
```

#### 3. Write a Pandas program to count the number of rows and columns of a DataFrame.

```python
# Count the number of rows and columns
rows, columns = df.shape

print(f"Number of rows: {rows}")
print(f"Number of columns: {columns}")
```

#### 4. Write a Pandas program that scatters Plot with Seaborn and Pandas.

```python
# Imporint Seaborn and matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame(data)

# Create a scatter plot using Seaborn
sns.scatterplot(x='Name', y='Age', data=df)

# Add labels and title
plt.xlabel('X-axis (Name)')
plt.ylabel('Y-axis (Age)')
plt.title('Scatter Plot Example')

# Show the plot
plt.show()
```

#### 5. Write a Pandas program to create a Pair Plot with Seaborn.

```python
# data = {
#     'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
#     'Age': [24, 27, 22, 32, 29],
#     'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
# }

df = pd.DataFrame(data)

df['City_encoded'] = pd.factorize(df['City'])[0]

# Create a pair plot with the 'Age' and encoded 'City' columns
sns.pairplot(df[['Age', 'City_encoded']])

plt.show()
```
