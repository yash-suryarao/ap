#### 1. Write a Python programming to display a bar chart of the popularity of programming Languages

```python
import matplotlib.pyplot as plt

languages = ['Python', 'Java', 'C', 'JavaScript', 'C++', 'Ruby', 'PHP']
popularity = [30, 10, 28, 5, 25, 12, 8]  # Hypothetical popularity values

plt.figure(figsize=(10, 6))
plt.bar(languages, popularity, color='skyblue')

# Adding title and labels
plt.title('Popularity of Programming Languages', fontsize=14)
plt.xlabel('Programming Languages', fontsize=12)
plt.ylabel('Popularity (%)', fontsize=12)

# Display the bar chart
plt.show()
```

#### 2. Write a Python programming to create a pie chart of the popularity of programming Languages.

```python
# Create the pie chart
plt.figure(figsize=(5, 5))
plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'orange', 'lightcoral', 'gold', 'lightpink', 'lightgray', 'lightyellow'])

# Adding a title
plt.title('Popularity of Programming Languages', fontsize=14)

# Display the pie chart
plt.show()
```

#### 3. Write a Python program to draw a scatter graph taking a random distribution in X and Y and plotted against each other.

```python
import numpy as np

# Generate random data for X and Y (e.g., 100 random values)
X = np.random.rand(100)
Y = np.random.rand(100)

plt.scatter(X, Y, color='blue', alpha=0.6)

plt.title('Random Distribution Scatter Plot', fontsize=14)
plt.xlabel('X values', fontsize=12)
plt.ylabel('Y values', fontsize=12)

plt.show()
```

#### 4. Write a Python program to draw a scatter plot with empty circles taking a random distribution in X and Y and plotted against each other.

```python
# Generate random data for X and Y (e.g., 100 random values)
X = np.random.rand(100)
Y = np.random.rand(100)

plt.scatter(X, Y, color='white', edgecolor='black', marker='o', s=100, alpha=0.8)

plt.title('Scatter Plot with Empty Circles', fontsize=14)
plt.xlabel('X values', fontsize=12)
plt.ylabel('Y values', fontsize=12)

plt.show()
```

#### 5. Write a Python program to draw a scatter plot using random distributions to generate balls of different sizes.

```python
# Generate random data for X and Y
X = np.random.rand(100)
Y = np.random.rand(100)

# Generate random sizes for the "balls"
sizes = np.random.uniform(50, 500, 100)

# Create the scatter plot with different sized balls
plt.scatter(X, Y, s=sizes, color='blue', alpha=0.6)

plt.title('Scatter Plot with Balls of Different Sizes', fontsize=14)
plt.xlabel('X values', fontsize=12)
plt.ylabel('Y values', fontsize=12)

plt.show()
```
